```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
plt.rc('font',family='serif', serif=['Palatino'])

CB91_Blue = '#2CBDFE'
CB91_Green = '#47DBCD'
CB91_Pink = '#F3A0F2'
CB91_Purple = '#9D2EC5'
CB91_Violet = '#661D98'
CB91_Amber = '#F5B14C'
color_list = [CB91_Blue, CB91_Green, CB91_Amber, CB91_Violet]
```

```python
x = [5, 10, 20, 50, 75, 100]
naive_h = [12.5, 9.9, 8.1, 3.1, 0.2, 0.2]
naive_h_error = np.asarray([8.1, 6.1, 5.4, 2.2, 0.4, 0.3])
informed_h = [5.8, 1.2, 0.7, 0.1, 0.1, 0.1]
informed_h_error = np.asarray([1.1, 0.2, 0.2, 0.1, 0.1, 0.1])
oracle = [3.1, 1.1, 0.2, 0.1, 0.1, 0.1]
oracle_error = np.asarray([0.9, 0.3, 0.1, 0.1, 0.0, 0.0])
```

```python
fig1 = plt.figure(figsize=(12,9))
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=color_list)
plt.legend(frameon=False)
ax1 = fig1.add_subplot()
ax1.set_title("Multi-Object Navigation (what failures)")
ax1.plot(x, naive_h, '--', label="Naive Human", marker='o', markersize=12)
ax1.fill_between(x, naive_h+naive_h_error, naive_h-naive_h_error,  alpha=0.2)
ax1.plot(x, informed_h, '--', label="Ours", marker='o', markersize=12)
ax1.fill_between(x, informed_h+informed_h_error, informed_h-informed_h_error, alpha=0.2)
ax1.plot(x, oracle, '--', label="Oracle", marker='o', markersize=12)
ax1.fill_between(x, oracle+oracle_error, oracle-oracle_error, alpha=0.2)

ax1.legend(prop={'size': 25})
for tick in ax1.get_xticklabels():
    tick.set_rotation(45)
ax1.set_ylabel('Distance from goal')
ax1.set_xlabel('Novel demonstrations generated')
for item in ([ax1.title, ax1.xaxis.label, ax1.yaxis.label] +
             ax1.get_xticklabels() + ax1.get_yticklabels()):
    item.set_fontsize(30)
fig1.savefig('what-demos.png')
```
