```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = [5, 10, 20, 50, 75, 100]
naive_h = [12.5, 9.9, 8.1, 3.1, 0.2, 0.2]
naive_h_error = np.asarray([8.1, 6.1, 5.4, 2.2, 0.4, 0.3])
informed_h = [5.8, 1.2, 0.7, 0.1, 0.1, 0.1]
informed_h_error = np.asarray([1.1, 0.2, 0.2, 0.1, 0.1, 0.1])
oracle = [3.1, 1.1, 0.2, 0.1, 0.1, 0.1]
oracle_error = np.asarray([0.9, 0.3, 0.1, 0.1, 0.0, 0.0])
```

```python
fig1 = plt.figure(figsize=(12,9))
ax1 = fig1.add_subplot()
ax1.set_title("Multi-Object Navigation (what failures)")
ax1.plot(x, naive_h, '--', color="orange", label="Naive Human", marker='o', markersize=12)
ax1.fill_between(x, naive_h+naive_h_error, naive_h-naive_h_error, color='orange', alpha=0.2)
ax1.plot(x, informed_h, '--', color="green", label="Informed Human", marker='o', markersize=12)
ax1.fill_between(x, informed_h+informed_h_error, informed_h-informed_h_error, color='green', alpha=0.2)
ax1.plot(x, oracle, '--', color="blue", label="Oracle", marker='o', markersize=12)
ax1.fill_between(x, oracle+oracle_error, oracle-oracle_error, color='blue', alpha=0.2)

ax1.legend(prop={'size': 25})
for tick in ax1.get_xticklabels():
    tick.set_rotation(45)
ax1.set_ylabel('Distance from goal')
ax1.set_xlabel('Number of new demonstrations generated')
for item in ([ax1.title, ax1.xaxis.label, ax1.yaxis.label] +
             ax1.get_xticklabels() + ax1.get_yticklabels()):
    item.set_fontsize(30)
fig1.savefig('what-demos.png')
```
